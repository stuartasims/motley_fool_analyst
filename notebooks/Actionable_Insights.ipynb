# A/B Test Recommendation

Because mobile device traffic accounts for the majority of sessions and tends to bounce very highly, optimizing the mobile experience on the most commonly-used devices / resolutions should decrease bounce rate. We can quantify this using Google Analytics bounce rate after ensuring device grid characteristics are stored.

The test could be as simple as taking an individual page, duplicating, and revising it to make it faster, break up large blocks of text with h2 tags, rework the keywords in meta data, linking to other site pages with relevant content, but leave the actual content identical to the original. 100% of user traffic could be included in the test with an even 50-50 split between the two pages.

With about 185,000 sessions per month on this site, we can create a conservative estimate of sessions at 80% of this value: 148,000 sessions. A sample size of 74,000 users (that same 50-50 split) would allow confident, statistical (detection of <1% change in the rate within one month. Higher change thresholds could reduce that time, but allowing the test to run across at least two business cycles is a best practice that would increase confidence in the results.
