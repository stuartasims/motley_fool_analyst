### Check for Missing Data

```python
for i in [traffic_source, geo_location, device, demos]:
    print(i.isnull().sum() / len(i))
    print("\n\n")
```

### Restructuring  datetime fields

I'm adding a column for month_year_dt which is a datetime type value capturing the month and year data. Two dataframes need this: traffic_source and device.

```python
# small data, can just use visual inspection to check for any missing values
print("Original month year (YYYYmm) field values: \n\n", traffic_source["Month of Year"].unique())

# create datetime field for plotting and other helpful dataframe operations later
traffic_source["month_year_dt"] = pd.to_datetime(
    traffic_source["Month of Year"], format="%Y%m")

print("\n\nAfter reformatting: \n\n", traffic_source.month_year_dt.unique())
```

### Adding Calculated Columns of Interest

```python
# small data, can just use visual inspection to check for any missing values
print("Original month year (YYYYmm) field values: \n\n", device["Month"].unique())

# create datetime field for plotting and other helpful dataframe operations later
device["month_year_dt"] = pd.to_datetime(
    device["Month"], format="%Y%m")

print("\n\nAfter reformatting: \n\n", device.month_year_dt.unique())
```

```python
# visualizing some metrics in terms of user volume may be helpful later

# by channel
traffic_source['bounced_sessions'] = np.round(traffic_source['Sessions'] * traffic_source['Bounce Rate'],0)
traffic_source['email_conv_sessions'] = np.round(traffic_source['Sessions'] * traffic_source['Email Capture Conversion Rate'],0)

# by geography
geo_location['bounced_sessions'] = np.round(geo_location['Sessions'] * geo_location['Bounce Rate'],0)
geo_location['email_conv_sessions'] = np.round(geo_location['Sessions'] * geo_location['Email Capture Conversion Rate'],0)
geo_location['total_duration'] = geo_location['Avg. Session Duration'] * geo_location['Sessions'] # estimate of time per country
geo_location['pct_overall_sessions'] = geo_location['Sessions'] / geo_location['Sessions'].sum()

# by device
device['bounced_sessions'] = np.round(device['Sessions'] * device['Bounce Rate'],0)
device['email_conv_sessions'] = np.round(device['Sessions'] * device['Email Capture Conversion Rate'],0)


```

### Exporting processed data

```python
# store list of dataframes for parsing
df_list = [traffic_source, geo_location, device, demos]
sheet_names = ['traffic_source', 'geo_location', 'device', 'demos']

writer = ExcelWriter('../data/processed/ga_export_processed.xlsx')
for n, df in enumerate(df_list):
    df.to_excel(writer, sheet_name = sheet_names[n], index=False)
    writer.save()
```
